## short cadence

```python
import tql

#uses first available sector and PDCSAP lightcurves by default
fig = tql.plot_tql(toiid=125)
```

```python
#specify sector and lightcurve type
fig = tql.plot_tql(toiid=125, sector=2, lctype='sap')
```

## long cadence

### custom aperture

```python
import tql

#specify long cadence and aperture shape
fig = tql.plot_tql(ticid=291555748, cadence='long', 
                   sap_mask='threshold', threshold_sigma=3,
                   sigma=(10,3)
                  )
```

### CDIPS pipeline lightcurve

```python
#See https://archive.stsci.edu/hlsp/cdips
fig = tql.plot_tql(ticid=88977253, cadence='long', lctype='cdips')
```

### PATHOS pipeline lightcurve

```python
#See https://archive.stsci.edu/hlsp/pathos
fig = tql.plot_tql(ticid=88977253, cadence='long', 
                   lctype='pathos', sigma=(10,3)
                  )
```
